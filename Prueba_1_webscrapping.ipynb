```python
import requests
import os
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import urllib
import pandas as pd
from selenium.webdriver.common.keys import Keys
```

```python
# URL de la página web a la que haremos scraping
url1 = "https://www.wikiart.org/en/paintings-by-style/art-brut?select=featured#!#filterName:featured,viewType:masonry"
url2 = "https://corbah.com/"
```

```python
driver = webdriver.Chrome()
driver.get(url1)
time.sleep(3)

img_tags = driver.find_elements(by=By.TAG_NAME, value="img")
directory = "images"

# Crear un directorio para almacenar las imágenes
if not os.path.exists("directory"):
    os.makedirs("directory")

# Iterar sobre las etiquetas 'img' y obtener las URLs de las imágenes
for img_tag in img_tags:
    img_url = img_tag.get_attribute("src")
    print(img_url)

    # Descargar la imagen y guardarla en el directorio creado
    if not img_url == None:
        img_name = img_url.split("/")[-1]
        img_path = os.path.join("directory", img_name)
        urllib.request.urlretrieve(img_url, img_path)
        print("Imagen descargada:", img_name)

# Cerrar el WebDriver
driver.quit()
```

```python
for img_tag in img_tags:
    img_url = img_tag.get_attribute("src")
    try:
        # Obtener el nombre de archivo de la URL
        img_name = img_url.split("/")[-1]
        # Ruta completa para guardar la imagen
        img_path = os.path.join(directory, img_name)

        
        # Descargar la imagen
        urllib.request.urlretrieve(img_url, img_path)
        print("Imagen descargada:", img_name)
    except Exception as e:
        print(f"Error al descargar la imagen {img_url}: {e}")

print("Todas las imágenes han sido descargadas.")
```

```python
driver = webdriver.Chrome()
driver.get(url1)
time.sleep(5)
for i in range(8):
   elemento_cargar_mas = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "load-more-phrase")))
   # Hacer clic en el botón de "CARGAR MÁS"
   elemento_cargar_mas.click() 
   time.sleep(5)

```

```python
time.sleep(3)
img_tags = driver.find_elements(by=By.TAG_NAME, value="img")
data = pd.DataFrame()
autor = []
img_nombre = []

for img_tag in img_tags:
    img_url = img_tag.get_attribute("src")
    try: 
        # Obtener el nombre de archivo de la URL
        img = img_url.split("/")[-1]
        autor_ = img_url.split("/")[-2]
        img_name = img.split(".")[0]
        # Ruta completa para guardar la imagen
        img_path = os.path.join(directory, img_name)
        img_nombre.append(img_name)
        autor.append(autor_)
        print(img_url)
    except Exception as e:
        print(f"Error None type: {e}")

print("Todas las imágenes han sido descargadas.")
```

```python
frases_concatenadas = []

for f1, f2 in zip(autor, img_nombre):
    concatenada = f1 + "_" + f2
    frases_concatenadas.append(concatenada)

print(frases_concatenadas)
data['painting'] = frases_concatenadas
```

```python
data['painting'].unique()
```

```python

data['address'] = 'https://uploads.wikiart.org/images/' + data['painting'].str.replace('_','/') + '.jpg'
```

```python
print(data.to_string(index=False, col_space=15, max_rows=50))
```

```python
# Carpeta de destino para guardar los archivos descargados
carpeta_destino = 'archivos_descargados'

# Crear la carpeta de destino si no existe
if not os.path.exists(carpeta_destino):
    os.makedirs(carpeta_destino)

# Descargar archivos 
for url in data['address']:
    #nombre del archivo del URL
    nombre_archivo = url.split('/')[-1]
    
    # Ruta completa del archivo en la carpeta de destino
    ruta_archivo = os.path.join(carpeta_destino, nombre_archivo)
    
    # Descargar el archivo
    response = requests.get(url)
    
    # Guardar el archivo descargado
    with open(ruta_archivo, 'wb') as archivo:
        archivo.write(response.content)
        print(f"Archivo descargado: {nombre_archivo}")

print("Descarga completa.")
```
