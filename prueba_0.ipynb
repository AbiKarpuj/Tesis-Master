```python
data = pd.read_csv('artemis_dataset_release_v0.csv')
```

```python
len(data['painting'].unique())
data['art_style'].unique()
```

```python
data['rand'] = data['painting'].rank(method='dense') 
```

```python
data['rand_int'] = ((data['rand'] % 9) + 1).astype(int).astype(str)
```

```python
data['address'] = 'https://uploads.wikiart.org/images/' + data['painting'].str.replace('_','/') + '.jpg'
```

```python
data['painting'].str.replace('_','/')[0]
```

```python
lista_estilos = data['art_style'].unique()
```

```python
#Generamos el formato para crear un a lista con las imagenes de cada estilo.
resultados = []

for estilo in lista_estilos:
    # Crea la cadena de texto correspondiente y agrégala a la lista de resultados
    resultados.append(f"{estilo} = data[data['art_style'] == '{estilo}']")

```

```python
# Imprime los resultados
for i in resultados:
    print(i)
resultados[5]
```

```python
Romanticism = data[data['art_style'] == 'Romanticism']
Romanticism
```

```python
# Carpeta de destino para guardar los archivos descargados
carpeta_destino = 'Romanticism'

# Crear la carpeta de destino si no existe
if not os.path.exists(carpeta_destino):
    os.makedirs(carpeta_destino)

# Descargar archivos 
for url in Romanticism['address']:
    #nombre del archivo del URL
    nombre_archivo = url.split('/')[-1]
    
    # Ruta completa del archivo en la carpeta de destino
    ruta_archivo = os.path.join( carpeta_destino, nombre_archivo)
    
    # Descargar el archivo
    response = requests.get(url)
    
    # Guardar el archivo descargado
    with open(ruta_archivo, 'wb') as archivo:
        archivo.write(response.content)
        print(f"Archivo descargado: {nombre_archivo}")

print("Descarga completa.")
```
